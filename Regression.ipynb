# Multi Linear Regression

```python
# extract features

x = house_data.loc[:, ['bathrooms', 'sqft_living', 'grade', 'sqft_above', 'sqft_living15']].values
```

```python
# Standardise the features

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
x_scaled
```

```python
from sklearn.decomposition import PCA
pc = PCA(n_components=3)
```

```python
p = pc.fit_transform(x_scaled)
p
```

```python
house_data1 = pd.DataFrame(data = p, columns= ['PC1','PC2','PC3'])
house_data_df = pd.concat([house_data1, house_data[['price']]], axis = 1)
house_data_df.head()
```

```python
# Multi linear regression

X=house_data_df[["PC1","PC2","PC3"]]  
y=house_data_df["price"].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=45) 
model=LinearRegression() 
model.fit(X_train,y_train) 
predicted=model.predict(X_test) 
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted))
```
